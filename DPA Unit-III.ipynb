# In the following program, the t-test is used to compare the means of customer spending between the two stores to determine whether there is a statistically significant difference between them.

Role of T-Test:
Hypothesis Testing: The t-test evaluates whether the difference in the average spending between Store 1 and Store 2 is significant or if it occurred by chance.

Null Hypothesis (H₀): There is no significant difference in the average spending between the two stores.

Alternative Hypothesis (H₁): There is a significant difference in the average spending between the two stores.

T-Statistic: It measures the size of the difference relative to the variation in the sample data. A higher absolute t-value suggests a greater difference.

P-Value: The p-value indicates the probability of obtaining the observed difference (or a more extreme one) if the null hypothesis is true.

If p-value < 0.05 (assuming a 5% significance level), you reject the null hypothesis, meaning the difference is statistically significant.

If p-value ≥ 0.05, you fail to reject the null hypothesis, meaning no significant difference was found.



```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data for two stores with 30 entries each
np.random.seed(42)
store1_amounts = np.random.randint(1000, 50000, 30)
store2_amounts = np.random.randint(1000, 50000, 30)

store1_df = pd.DataFrame({'Customer_ID': range(1, 31), 'Amount': store1_amounts})
store2_df = pd.DataFrame({'Customer_ID': range(1, 31), 'Amount': store2_amounts})

# Save data to CSV files
store1_df.to_csv('store1_data.csv', index=False)
store2_df.to_csv('store2_data.csv', index=False)
print("CSV files created: store1_data.csv and store2_data.csv")

# Read data from CSV files
store1_data = pd.read_csv('store1_data.csv')['Amount']
store2_data = pd.read_csv('store2_data.csv')['Amount']

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(store1_data, store2_data)

# Display results
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

# Visualization
plt.figure(figsize=(8, 6))
sns.histplot(store1_data, kde=True, color='blue', label='Store 1')
sns.histplot(store2_data, kde=True, color='orange', label='Store 2')
plt.title('Customer Spending Comparison Between Two Stores')
plt.xlabel('Amount Spent')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# In the following program, the ANOVA (Analysis of Variance) test is used to determine whether there are statistically significant differences in the average customer spending among the three banks.

Role of ANOVA Test:
Comparing Multiple Groups: ANOVA is particularly useful when comparing the means of three or more independent groups (in this case, Bank 1, Bank 2, and Bank 3) to see if there is a significant difference.

Hypothesis Testing: The ANOVA test uses the following hypotheses:

Null Hypothesis (H₀): The means of customer spending across all three banks are equal (no significant difference).

Alternative Hypothesis (H₁): At least one bank has a significantly different mean customer spending.

F-Statistic: ANOVA calculates an F-statistic, which measures the ratio of variation between the group means to the variation within each group. A higher F-statistic indicates a larger difference between group means.

P-Value: The p-value indicates the probability of observing the differences by random chance.

If p-value < 0.05 (assuming a 5% significance level), the null hypothesis is rejected, meaning a significant difference exists.

If p-value ≥ 0.05, the null hypothesis is not rejected, indicating no significant difference.

Note:
If the ANOVA test detects a significant difference, further analysis (like post hoc tests) can be performed to identify which specific banks differ in customer spending.

The test provides valuable insights for bank management to understand customer spending patterns and implement appropriate financial strategies.

```python
import numpy as np
import pandas as pd
from scipy import stats

# Generate sample data for three banks with 30 entries each
np.random.seed(42)
bank1_amounts = np.random.randint(1000, 50000, 30)
bank2_amounts = np.random.randint(1000, 50000, 30)
bank3_amounts = np.random.randint(1000, 50000, 30)

bank1_df = pd.DataFrame({'Customer_ID': range(1, 31), 'Amount': bank1_amounts})
bank2_df = pd.DataFrame({'Customer_ID': range(1, 31), 'Amount': bank2_amounts})
bank3_df = pd.DataFrame({'Customer_ID': range(1, 31), 'Amount': bank3_amounts})

# Save data to CSV files
bank1_df.to_csv('bank1_data.csv', index=False)
bank2_df.to_csv('bank2_data.csv', index=False)
bank3_df.to_csv('bank3_data.csv', index=False)
print("CSV files created: bank1_data.csv, bank2_data.csv, and bank3_data.csv")

# Read data from CSV files
bank1_data = pd.read_csv('bank1_data.csv')['Amount']
bank2_data = pd.read_csv('bank2_data.csv')['Amount']
bank3_data = pd.read_csv('bank3_data.csv')['Amount']

# Perform one-way ANOVA test
f_stat, p_value = stats.f_oneway(bank1_data, bank2_data, bank3_data)

# Display results
print(f"F-Statistic: {f_stat}")
print(f"P-Value: {p_value}")

# Conclusion
if p_value < 0.05:
    print("There is a significant difference in customer spending among the three banks.")
else:
    print("There is no significant difference in customer spending among the three banks.")
```

# In the following program, the F-test is used to compare the variances of customer spending between the two banks. 

Role of the F-Test:
- Variance Comparison: The F-test checks whether the variances (a measure of spread) of the two datasets are significantly different. It is commonly used to test the assumption of equal variances, which is often a requirement for other statistical tests like the t-test.
  
- Hypothesis Testing: The F-test follows these hypotheses:
  - Null Hypothesis (H₀): The variances of customer spending between Bank 1 and Bank 2 are equal.
  - Alternative Hypothesis (H₁): The variances of customer spending between Bank 1 and Bank 2 are significantly different.

F-Statistic Calculation:  
  The F-statistic is calculated as:  
  \[
  F = \frac{\text{Variance of Bank 1 Data}}{\text{Variance of Bank 2 Data}}
  \]
  - If the F-statistic is significantly greater or smaller than 1, it suggests unequal variances.
  
- P-Value Interpretation:
  - If p-value < 0.05 (assuming a 5% significance level), the null hypothesis is rejected, meaning there is a significant difference in the variances.
  - If p-value ≥ 0.05, the null hypothesis is not rejected, indicating no significant difference in the variances.
Note:
- The F-test in this program helps assess whether the spread of customer spending is consistent between the two banks.
- If a significant difference is found, it may indicate varying spending behaviors or differences in the customer base between the banks.

```python
import numpy as np
import pandas as pd
from scipy import stats

# Generate sample data for two banks with 30 entries each
np.random.seed(42)
bank1_amounts = np.random.randint(1000, 50000, 30)
bank2_amounts = np.random.randint(1000, 50000, 30)

bank1_df = pd.DataFrame({'Customer_ID': range(1, 31), 'Amount': bank1_amounts})
bank2_df = pd.DataFrame({'Customer_ID': range(1, 31), 'Amount': bank2_amounts})

# Save data to CSV files
bank1_df.to_csv('bank1_data.csv', index=False)
bank2_df.to_csv('bank2_data.csv', index=False)
print("CSV files created: bank1_data.csv and bank2_data.csv")

# Read data from CSV files
bank1_data = pd.read_csv('bank1_data.csv')['Amount']
bank2_data = pd.read_csv('bank2_data.csv')['Amount']

# Perform F-Test to compare variances
f_stat = np.var(bank1_data, ddof=1) / np.var(bank2_data, ddof=1)
df1 = len(bank1_data) - 1
df2 = len(bank2_data) - 1
p_value = stats.f.cdf(f_stat, df1, df2)

# Display results
print(f"F-Statistic: {f_stat}")
print(f"P-Value: {2 * min(p_value, 1 - p_value)}")

# Conclusion
if p_value < 0.05:
    print("There is a significant difference in variances between the two banks.")
else:
    print("There is no significant difference in variances between the two banks.")
```

# In the following program, the correlation coefficient measures the strength and direction of the linear relationship between the **Product_Price** and **Discount** columns.

Role of Correlation Coefficient:
-Quantifies Relationship: It provides a numerical value (ranging from -1 to +1) that quantifies how closely the two variables are related.
  
- Interpretation:
  - +1: Perfect positive correlation (when one variable increases, the other increases).
  - -1: Perfect negative correlation (when one variable increases, the other decreases).
  - 0: No correlation (no linear relationship between the variables).
  
- Decision-Making Insight:
  - A positive correlation indicates that higher-priced products tend to have higher discounts.
  - A negative correlation means higher-priced products tend to have lower discounts.
  - A near-zero correlation suggests no meaningful relationship between product price and discount.

- Visualization Support: The scatter plot and the regression line (red line) visually support the correlation result, making it easier to interpret the relationship between the two variables.

In short, the correlation coefficient helps in understanding pricing strategies and discount patterns in the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data for sales dataset with Product Price and Discount
np.random.seed(42)
product_price = np.random.randint(100, 1000, 50)
discount = np.random.randint(5, 100, 50)

sales_df = pd.DataFrame({
    'Product_Price': product_price,
    'Discount': discount
})

# Save data to CSV file
sales_df.to_csv('sales_data.csv', index=False)
print("CSV file created: sales_data.csv")

# Read data from CSV file
data = pd.read_csv('sales_data.csv')

# Perform correlation analysis
correlation_coefficient = data['Product_Price'].corr(data['Discount'])

# Display result
print(f"Correlation Coefficient between Product Price and Discount: {correlation_coefficient}")

# Conclusion
if correlation_coefficient > 0:
    print("There is a positive correlation between Product Price and Discount.")
elif correlation_coefficient < 0:
    print("There is a negative correlation between Product Price and Discount.")
else:
    print("There is no correlation between Product Price and Discount.")
    


# Visualization using scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Product_Price', y='Discount', data=data)
sns.regplot(x='Product_Price', y='Discount', data=data, scatter=False, color='red')
plt.title(f'Scatter Plot with Correlation Coefficient: {correlation_coefficient:.2f}')
plt.xlabel('Product Price')
plt.ylabel('Discount')
plt.grid(True)
plt.show()
```

# The following program performs a Linear Regression Analysis to determine the relationship between Product Price and Discount in the sales dataset. 

Key Significance of the Program:
1. Regression Coefficient (Slope):
    - The slope measures how much the Discount changes for every unit change in Product Price.
    - A positive slope indicates that as the product price increases, the discount tends to increase.
    - A negative slope suggests that higher-priced products receive lower discounts.
    - A slope of zero means no relationship between the product price and discount.

2. Intercept:
    - The intercept is the value of the discount when the product price is zero. 
    - While it may not always have a practical interpretation, it helps in forming the linear equation:
      \[
      \text{Discount} = (\text{Regression Coefficient} \times \text{Product Price}) + \text{Intercept}
      \]

3. Decision Making:
    - Businesses can analyze pricing strategies by understanding how discounts are applied based on product prices.
    - If the slope is significantly negative or positive, businesses may adjust their discounting policies accordingly.
    - It can also help in predicting discounts for future product pricing.

Overall, this program provides insights into how discounts are influenced by product prices and supports strategic pricing decisions. 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Read data from the sales dataset CSV file
data = pd.read_csv('sales_data.csv')

# Perform Linear Regression Analysis
X = data[['Product_Price']]
y = data['Discount']

model = LinearRegression()
model.fit(X, y)

# Compute Regression Coefficient and Intercept
regression_coefficient = model.coef_[0]
intercept = model.intercept_

print(f"Regression Coefficient (Slope): {regression_coefficient}")
print(f"Intercept: {intercept}")
```

# The following program implements K-Means Clustering to segment data into groups based on the relationship between Product Price and Discount.

Significance of the Program:

1. Data Segmentation:  
    - The program divides the dataset into 3 clusters (as specified) using the K-Means algorithm.
    - Each cluster contains data points that are similar to each other based on their product price and discount.

2. K-Means Algorithm:  
    - It is an unsupervised machine learning algorithm used for clustering tasks.
    - The algorithm finds groups by minimizing the sum of squared distances between data points and their assigned cluster center (centroid).
  
3. Standardization: 
    - Data is scaled using StandardScaler to ensure that both columns (Product Price and Discount) contribute equally to the clustering process.
  
4. Cluster Visualization:  
    - The program uses Matplotlib to plot the data points and visualize how they are grouped.
    - Different colors represent different clusters.
    - Red X markers indicate the centroids (the central points of the clusters).

5. Business Insights:  
    - Businesses can use this clustering to identify pricing patterns and discount strategies.
    - For example:
      - High Price, Low Discount products could be luxury items.
      - Low Price, High Discount products may be promotional items.
      - Moderate Price, Moderate Discount could represent standard goods.
  
6. Further Applications:This analysis can help in personalized marketing, customer segmentation, and targeted promotions.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Generate sample data for clustering
np.random.seed(42)
product_price = np.random.randint(100, 1000, 100)
discount = np.random.randint(5, 100, 100)

sales_df = pd.DataFrame({
    'Product_Price': product_price,
    'Discount': discount
})

# Standardize the data
scaler = StandardScaler()
sales_scaled = scaler.fit_transform(sales_df)

# Perform K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
sales_df['Cluster'] = kmeans.fit_predict(sales_scaled)

# Visualizing the clusters
plt.figure(figsize=(8, 6))
plt.scatter(sales_df['Product_Price'], sales_df['Discount'], c=sales_df['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0] * scaler.scale_[0] + scaler.mean_[0],
            kmeans.cluster_centers_[:, 1] * scaler.scale_[1] + scaler.mean_[1],
            s=300, c='red', marker='X', label='Centroids')

plt.title('K-Means Clustering of Product Price and Discount')
plt.xlabel('Product Price')
plt.ylabel('Discount')
plt.legend()
plt.grid(True)
plt.show()
```

# The following program implements Hierarchical Clustering to group data points based on the similarity between their Product Price and Discount.


1. Hierarchical Clustering Overview: 
    - Hierarchical clustering is an unsupervised machine learning technique used to create a tree-like structure (dendrogram) that visualizes how data points are grouped together.
    - It does not require specifying the number of clusters in advance, unlike K-Means.

2. Linkage Method:  
    - The program uses the Ward’s method for linkage, which minimizes the variance within clusters.
    - It evaluates the distance between clusters based on their centroids and merges them to reduce overall variance.

3. Dendrogram Visualization:  
    - The dendrogram is a visual representation of the clustering process.
    - The y-axis represents the Euclidean distance between clusters.
    - The x-axis shows the individual data points.
    - The longer the vertical lines, the greater the difference between clusters.

4. Interpretation from Dendrogram:
    - You can cut the dendrogram at a desired height to form clusters.
    - Large gaps between horizontal lines suggest clear separations between clusters.
    - Helps in understanding how similar or different products are based on their price and discount patterns.

5. Business Insights: 
    - Businesses can analyze pricing and discount strategies by identifying clusters of products with similar behaviors.
    - It can aid in decisions such as promotional offers, inventory management, or personalized discounts.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# Generate sample data for clustering
np.random.seed(42)
product_price = np.random.randint(100, 1000, 50)
discount = np.random.randint(5, 100, 50)

sales_df = pd.DataFrame({
    'Product_Price': product_price,
    'Discount': discount
})

# Standardize the data
scaler = StandardScaler()
sales_scaled = scaler.fit_transform(sales_df)

# Perform Hierarchical Clustering
linked = linkage(sales_scaled, method='ward')

# Plot Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, labels=sales_df.index, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.grid(True)
plt.show()
```

# The following program implements DBSCAN (Density-Based Spatial Clustering of Applications with Noise) to cluster data points based on their density in the **Product Price** and **Discount** dataset.

Significance of the Program:

1. DBSCAN Overview: 
    - DBSCAN is an unsupervised machine learning algorithm used for clustering datasets based on density.
    - It groups together data points that are closely packed (high-density areas) and marks points in low-density regions as noise or outliers.

2. Key Parameters Used:  
    - `eps=0.5`: The maximum distance between two points for one to be considered in the neighborhood of the other.
    - `min_samples=5`: The minimum number of data points required to form a dense region (cluster).
  
3. Noise and Outlier Detection:  
    - Unlike K-Means or Hierarchical Clustering, DBSCAN can detect and label outliers as -1 (noise).
    - This is useful for detecting anomalies in datasets.

4. Visualization of Clusters: 
    - The scatter plot uses Seaborn to visualize the clusters formed by DBSCAN.
    - Different clusters are shown in distinct colors, and outliers are typically labeled as Cluster -1.
  
5. Business Insights:  
    - Businesses can identify segments of products that have similar pricing and discount patterns.
    - Detecting outliers can help in identifying unusual pricing or promotional strategies.
    - Useful for fraud detection or customer segmentation.
Note: 
The program provides a clear visualization of how products are grouped based on their pricing and discounts, while also highlighting any anomalies in the data. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate sample data for clustering
np.random.seed(42)
product_price = np.random.randint(100, 1000, 100)
discount = np.random.randint(5, 100, 100)

sales_df = pd.DataFrame({
    'Product_Price': product_price,
    'Discount': discount
})

# Standardize the data
scaler = StandardScaler()
sales_scaled = scaler.fit_transform(sales_df)

# Perform DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
sales_df['Cluster'] = dbscan.fit_predict(sales_scaled)

# Visualizing the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Product_Price', y='Discount', hue='Cluster', data=sales_df, palette='viridis', legend='full')
plt.title('DBSCAN Clustering of Product Price and Discount')
plt.xlabel('Product Price')
plt.ylabel('Discount')
plt.grid(True)
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# The following program implements the Apriori Algorithm to identify frequent itemsets and generate association rules from transactional data.

1. Apriori Algorithm Overview:  
    - The Apriori Algorithm is a widely used unsupervised machine learning technique for association rule mining.
    - It identifies combinations of items that frequently occur together in transactions.
    - Typically applied in market basket analysis to understand customer purchasing behavior.

2. Frequent Itemsets Identification: 
    - The program uses the Apriori algorithm to find all item combinations that meet a specified minimum support threshold (`min_support=0.4`).
    - Support measures how frequently an itemset appears in the dataset.
  
3. Association Rule Generation:  
    - After identifying frequent itemsets, the program generates association rules using the `association_rules()` function.
    - Association rules help in finding patterns like:
      - "If a customer buys Bread, they are likely to buy Butter."**
    - Metrics used for rule evaluation include:
      - Support: Frequency of the itemset in all transactions.
      - Confidence: Probability that the consequent is purchased given the antecedent.
      - Lift: Strength of the rule compared to random chance.

4. Business Insights:  
    - Businesses can apply these rules for:
      - Product bundling: Offer discounts on items frequently bought together.
      - Recommendation systems: Suggest related products to customers.
      - Inventory management: Stock frequently bought items together.
  
Note:  
The program provides valuable insights into consumer behavior by identifying which products are often purchased together, enabling data-driven business strategies.


```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Sample transaction data
data = {'Transaction': [1, 2, 3, 4, 5],
        'Milk': [True, False, True, True, False],
        'Bread': [True, True, True, False, True],
        'Butter': [False, True, True, True, False],
        'Eggs': [True, True, False, True, True]}

df = pd.DataFrame(data)
df.set_index('Transaction', inplace=True)

# Applying Apriori Algorithm
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)

# Generating Association Rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

# The following program implements the   Eclat (Equivalence Class Transformation)   algorithm to find frequent itemsets in transactional data. Unlike the   Apriori   algorithm, which uses a breadth-first search,   Eclat   uses a depth-first search approach to efficiently discover frequent itemsets.

Significance of the Program:  

1.   Eclat Algorithm Overview:    
    -   Eclat   is a fast and memory-efficient algorithm for   frequent itemset mining  .
    - It works by recursively exploring all possible combinations of items (itemsets) and calculating their   support  .
    - Support is the number of transactions that contain a particular itemset.

2.   How the Program Works:    
    - The input data consists of binary transaction records, where `True` indicates the presence of an item in a transaction.
    - Transactions are converted into   sets   of purchased items for easier manipulation.
    - The   eclat()   function recursively generates all possible item combinations using depth-first search.
    - Only itemsets with a support greater than or equal to the specified   min_support   are considered   frequent  .

3.   Support Calculation:    
    - The support of an itemset is calculated as the number of transactions that contain that itemset.
    - The program uses `set.issubset()` to check if a particular itemset exists within each transaction.

4.   Applications of Eclat Algorithm:    
    -   Market Basket Analysis  : Identify frequently purchased product combinations for promotions and bundling.
    -   Recommendation Systems  : Suggest products to customers based on frequent itemsets.
    -   Inventory Management  : Optimize stock by understanding purchase patterns.
  
5.   Efficiency of Eclat:    
    - Eclat is particularly useful for datasets with a large number of frequent itemsets due to its efficient memory management.
    - It often outperforms Apriori in scenarios where memory is limited.



```python
import pandas as pd
from itertools import combinations

# Sample transaction data
data = {'Transaction': [1, 2, 3, 4, 5],
        'Milk': [True, False, True, True, False],
        'Bread': [True, True, True, False, True],
        'Butter': [False, True, True, True, False],
        'Eggs': [True, True, False, True, True]}

df = pd.DataFrame(data)
df.set_index('Transaction', inplace=True)

# Convert data to a list of sets for Eclat algorithm
transactions = []
for index, row in df.iterrows():
    transactions.append(set(row[row == True].index))

# Function to find frequent itemsets using Eclat
def eclat(prefix, items, min_support):
    frequent_itemsets = []
    for i in range(len(items)):
        item = items[i]
        new_prefix = prefix + [item]
        new_items = items[i+1:]
        support = sum(1 for transaction in transactions if set(new_prefix).issubset(transaction))
        if support >= min_support:
            frequent_itemsets.append((new_prefix, support))
            frequent_itemsets.extend(eclat(new_prefix, new_items, min_support))
    return frequent_itemsets

# Run Eclat with minimum support of 2
min_support = 2
unique_items = sorted(set(item for transaction in transactions for item in transaction))
frequent_itemsets = eclat([], unique_items, min_support)

# Display results
print("Frequent Itemsets:")
for itemset, support in frequent_itemsets:
    print(f"{itemset} - Support: {support}")
```

# The following program implements the   FP-Growth (Frequent Pattern Growth)   algorithm to identify frequent itemsets and generate association rules from transactional data.

1.   FP-Growth Algorithm Overview:    
    -   FP-Growth   is an   unsupervised machine learning   algorithm used for   frequent itemset mining  .
    - It is a more efficient alternative to the   Apriori Algorithm   since it doesn't generate candidate itemsets for every possible combination.
    - Instead, it constructs a   Frequent Pattern Tree (FP-Tree)   to compress the dataset by storing only frequent patterns.

2.   Key Steps in the Program:    
    -   Data Preparation:    
      The input dataset contains binary transaction data (`True` for purchased items and `False` for non-purchased items).
    -   FP-Growth Application:    
      - The algorithm identifies itemsets that occur frequently based on a specified   minimum support threshold   (`min_support=0.4`).
    -   Association Rule Generation:    
      - The program uses the `association_rules()` function to generate meaningful rules.
      - These rules are evaluated using metrics such as:
        -   Support:   Frequency of the itemset in all transactions.
        -   Confidence:   Likelihood that a consequent item is purchased if the antecedent is purchased.
        -   Lift:   Strength of the rule compared to random occurrence.

3.   Efficiency of FP-Growth:    
    - Unlike   Apriori  , FP-Growth scans the data only twice, making it faster and memory-efficient for large datasets.
    - It is particularly effective when there are many frequent itemsets.

4.   Business Insights:    
    -   Market Basket Analysis:   Identify products often bought together for promotional strategies.
    -   Recommendation Systems:   Suggest related products to customers using association rules.
    -   Inventory Management:   Optimize stock by identifying popular product bundles.
  


```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Sample transaction data
data = {'Transaction': [1, 2, 3, 4, 5],
        'Milk': [True, False, True, True, False],
        'Bread': [True, True, True, False, True],
        'Butter': [False, True, True, True, False],
        'Eggs': [True, True, False, True, True]}

df = pd.DataFrame(data)
df.set_index('Transaction', inplace=True)

# Applying FP-Growth Algorithm
frequent_itemsets = fpgrowth(df, min_support=0.4, use_colnames=True)
print("Frequent Itemsets:")
print(frequent_itemsets)

# Generating Association Rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
